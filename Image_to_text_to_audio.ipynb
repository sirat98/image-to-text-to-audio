```python
#Tesseract-OCR : is an open source optical character recognition engine it basically extract text from images and screenshots
#pytesseract is a python wrapper for tesseract-OCR
#gTTS stands for Google Text-to-Speech, which is a Python library


!sudo apt install tesseract-ocr
!pip install pytesseract
!pip install gTTS
!pip install Pillow==9.0.0
```

```python
import pytesseract
from PIL import Image
import gtts
```

```python
img = Image.open('/content/python.png')
print(img)
```

```python
result = pytesseract.image_to_string(img)
print(result)
tts = gtts.gTTS(result)
tts.save("hey.mp3")
```
